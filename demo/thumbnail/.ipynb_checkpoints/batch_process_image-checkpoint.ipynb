```python
from PIL import Image
import numpy as np
```

```python
img = Image.open('test_single.png')
data = np.asarray(img)
display(img)
```

```python
new_data = data.copy()

white = np.ones(3)*255
grey = np.ones(3)*232
off_white_0 = [254, 254, 255]
off_white_1 = [254, 254, 254]
off_grey_0 = [231, 231, 232]
off_grey_2 = [231. 231. 231]

g1 = [off_white_1]
g2 = [off_grey_0, off_grey_1]

for w, g in zip(g1, g2):
    to_grey = np.all(data==w, axis=2)
    to_white = np.all(data==g, axis=2)
    new_data[to_grey] = grey
    new_data[to_white] = white

new_data[np.all(data==grey, axis=2)] = white
new_data[np.all(data==white, axis=2)] = grey
    
new_img = Image.fromarray(new_data)
display(new_img)
```

```python
new_img.save('test.bmp', format='BMP')
```
